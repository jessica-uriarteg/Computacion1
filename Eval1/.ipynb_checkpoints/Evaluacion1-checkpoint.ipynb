```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
import statsmodels.api as sm
import scipy.stats as stats
from pylab import *
from scipy.io import loadmat
df=pd.read_csv("GLB.Ts+dSST.csv",skiprows=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
pd.to_numeric(df["J-D"], errors='coerce')
```

```python
pd.to_numeric(df["Year"], errors='coerce')
```

```python
plt.title("Global mean estimates based on land and ocean data")
plt.ylabel('Temperature Anomaly')

x=df["Year"]
y=df["J-D"]

plt.plot(x,y,marker='.',label='Annual Mean')
plt.plot(x,y,'k',label='_nolegend_')
lowess = sm.nonparametric.lowess(y,x,frac=0.1)
plt.plot(lowess[:,0],lowess[:,1],'r-',label='Lowess Smoothing')
plt.legend(loc='upper left')

plt.show()
```

```python
figure(1)
title('Variabilidad de temperatura promedio de cada mes')
df.boxplot(['Jan','Feb','Mar','Apr','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
xlabel=('Meses')
plt.show()
```
